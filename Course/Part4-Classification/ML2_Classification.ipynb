```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('health_data.csv')
```

```python
df = df[['weight','height','gender']]
ax = df[df['gender'] == 0].plot(x='weight',y='height',kind = 'scatter', color='r', alpha =0.5)
ax = df[df['gender'] == 1].plot(x='weight',y='height',kind = 'scatter', color='b', ax = ax, alpha=0.5)
```

```python
import numpy as np
np.random.rand()
```

```python
df['rand_num'] = df['gender'].apply(lambda x: np.random.rand())
test_index = int(len(df)*0.3)
test_df = df.sort_values('rand_num')[:test_index]
train_df = df.sort_values('rand_num')[test_index:]
```

```python
test_df['gender'].value_counts()
train_df['gender'].value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)
```

```python
clf.fit(train_df[['weight','height']], train_df['gender'])
```

```python
test_df['gender_pred'] = clf.predict(test_df[['weight','height']])
train_df['gender_pred'] = clf.predict(train_df[['weight','height']])
```

```python
def get_conf_matrix_df(test_df):
  TP = test_df[(test_df['gender'] == test_df['gender_pred'])&(test_df['gender'] == 1)]
  TN = test_df[(test_df['gender'] == test_df['gender_pred'])&(test_df['gender'] == 0)]
  FP = test_df[(test_df['gender'] != test_df['gender_pred'])&(test_df['gender_pred'] == 1)]
  FN = test_df[(test_df['gender'] != test_df['gender_pred'])&(test_df['gender_pred'] == 0)]
  return TP, TN, FP, FN
```

```python
TP, TN, FP, FN = get_conf_matrix_df(test_df)
print(f'TP:{len(TP)}, TN:{len(TN)}, FP:{len(FP)}, FN:{len(FN)}')

TP, TN, FP, FN = get_conf_matrix_df(train_df)
print(f'TP:{len(TP)}, TN:{len(TN)}, FP:{len(FP)}, FN:{len(FN)}')
```
